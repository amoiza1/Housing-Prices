```python
import pandas as pd
sample_submission = pd.read_csv("../input/house-prices-advanced-regression-techniques/sample_submission.csv")
test = pd.read_csv("../input/house-prices-advanced-regression-techniques/test.csv")
train = pd.read_csv("../input/house-prices-advanced-regression-techniques/train.csv")
```

```python
import numpy as np
y_nan=pd.DataFrame(np.repeat(np.nan,1459))
y_nan.columns=['SalePrice']
test_tm=pd.concat([test,y_nan],axis=1)
test_tm.head()
```

```python
all_data=train.append(test_tm)
```

```python
all_data['MasVnrType']=all_data.MasVnrType.fillna('None') 
all_data['MasVnrArea']=all_data.MasVnrArea.fillna(all_data.MasVnrArea.median()) 
```

```python
all_data['BsmtQual']=all_data.BsmtQual.fillna('None') 
all_data['BsmtCond']=all_data.BsmtCond.fillna('None') 
all_data['BsmtExposure']=all_data.BsmtExposure.fillna('None') 
all_data['BsmtFinType1']=all_data.BsmtFinType1.fillna('None') 
all_data['BsmtFinType2']=all_data.BsmtFinType2.fillna('None') 
```

```python
all_data['BsmtFinSF1']=all_data.BsmtFinSF1.fillna(all_data.BsmtFinSF1.median()) 
all_data['BsmtFinSF2']=all_data.BsmtFinSF2.fillna(all_data.BsmtFinSF2.median()) 
all_data['BsmtUnfSF']=all_data.BsmtUnfSF.fillna(all_data.BsmtUnfSF.median()) 
all_data['TotalBsmtSF']=all_data.TotalBsmtSF.fillna(all_data.TotalBsmtSF.median()) 
all_data['BsmtFullBath']=all_data.BsmtFullBath.fillna(all_data.BsmtFullBath.median()) 
all_data['BsmtHalfBath']=all_data.BsmtHalfBath.fillna(all_data.BsmtHalfBath.median()) 
```

```python
all_data['FireplaceQu']=all_data.FireplaceQu.fillna('None')
all_data['GarageType']=all_data.GarageType.fillna('None') 
all_data['GarageFinish']=all_data.GarageFinish.fillna('None') 
all_data['GarageQual']=all_data.GarageQual.fillna('None') 
all_data['GarageCond']=all_data.GarageCond.fillna('None') 
```

```python
all_data['GarageYrBlt']=all_data.GarageYrBlt.fillna(all_data.GarageYrBlt.median()) 
all_data['GarageCars']=all_data.GarageCars.fillna(all_data.GarageCars.median()) 
all_data['GarageArea']=all_data.BsmtHalfBath.fillna(all_data.GarageArea.median()) 
```

```python
all_data['PoolQC']=all_data.PoolQC.fillna('None') 
all_data['Fence']=all_data.Fence.fillna('None') 
all_data['MiscFeature']=all_data.MiscFeature.fillna('None') 
all_data['SaleType']=all_data.SaleType.fillna('Oth') 
```

```python
all_data['KitchenQual']=all_data.KitchenQual.fillna('None') 
all_data['Functional']=all_data.Functional.fillna('None') 
all_data['Electrical']=all_data.Electrical.fillna('None') 
```

```python
all_data.info()
```

```python
all_data_cat=all_data.loc[:,all_data.dtypes==object]
all_data_num=all_data.loc[:,all_data.dtypes!=object]
print(all_data_cat.shape)
print(all_data_num.shape)
```

```python
all_data_cat_dummies=pd.get_dummies(all_data_cat,drop_first=True)
all_data_cat_dummies
```

```python
all_data_num_filter=all_data_num.iloc[:,1:37]
all_data_num_filter
```

```python
from sklearn.preprocessing import StandardScaler
scaler=StandardScaler()
all_data_num_scaled=scaler.fit_transform(all_data_num_filter)
all_data_num_scaled=pd.DataFrame(all_data_num_scaled)
all_data_num_scaled.columns=all_data_num_filter.columns
all_data_num_scaled
```

```python
final_data=pd.concat([all_data_num_scaled.reset_index(drop=True),all_data_cat_dummies.reset_index(drop=True)],axis=1)
final_data
#all_data_num_scaled.shape
```

```python
X_train = final_data.iloc[:1460,:]
y_train = train.loc[:,['SalePrice']] 
X_test = final_data.iloc[1460:2919,:]
```

```python
print(X_train.shape)
print(y_train.shape)
print(X_test.shape)
```

```python
#Correlations
housing_corr=pd.concat([X_train,y_train],axis=1).corr()
housing_corr['SalePrice'].sort_values(ascending=False).head(20)
```

```python
#Linear Regression
from sklearn.linear_model import LinearRegression
lin_reg=LinearRegression()
lin_reg.fit(X_train,y_train)
```

```python
from sklearn.metrics import mean_squared_error
y_pred=lin_reg.predict(X_train)
lin_mse=mean_squared_error(y_train,y_pred)
lin_rmse=np.sqrt(lin_mse)
lin_rmse
```

```python
#Decision Tree
from sklearn.tree import DecisionTreeRegressor
tree_reg=DecisionTreeRegressor()
tree_reg.fit(X_train,y_train)
```

```python
y_pred=tree_reg.predict(X_train)
tree_mse=mean_squared_error(y_train,y_pred)
tree_rmse=np.sqrt(tree_mse)
tree_rmse
```

```python
#CV on Decision Trees
from sklearn.model_selection import cross_val_score
tree_scores=cross_val_score(tree_reg,X_train,y_train,scoring='neg_mean_squared_error',cv=10)
tree_rmse_scores=np.sqrt(-tree_scores)
tree_rmse_scores.mean()
```

```python
#CV on Linear Regression
from sklearn.model_selection import cross_val_score
scores=cross_val_score(lin_reg,X_train,y_train,scoring='neg_mean_squared_error',cv=10)
lin_rmse_scores=np.sqrt(-scores)
lin_rmse_scores.mean()
```

```python
from sklearn.ensemble import RandomForestRegressor
forest_reg=RandomForestRegressor()
forest_reg.fit(X_train,y_train)
```

```python
y_pred=forest_reg.predict(X_train)
forest_rmse=np.sqrt(mean_squared_error(y_train,y_pred))
forest_rmse
```

```python
forest_scores=cross_val_score(forest_reg,X_train,y_train,scoring='neg_mean_squared_error',cv=5)
forest_rmse=np.sqrt(-forest_scores)
forest_rmse.mean()
```

```python
y_test=forest_reg.predict(X_test)
y_test
```

```python
#PCA Try
from sklearn.decomposition import PCA
pca=PCA()
pca.fit(X_train)
cumsum=np.cumsum(pca.explained_variance_ratio_)
d=np.argmax(cumsum>=0.95)+1
pca2=PCA(n_components=d)
X_reduced=pca2.fit_transform(X_train)
```

```python
from sklearn.ensemble import GradientBoostingRegressor
grad_reg=GradientBoostingRegressor()
grad_reg.fit(X_reduced,y_train)
grad_score=cross_val_score(grad_reg,X_reduced,y_train,scoring='neg_mean_squared_error',cv=10)
grad_rmse=np.sqrt(-grad_score)
grad_rmse.mean()
```
